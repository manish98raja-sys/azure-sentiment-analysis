```python
from azure.ai.textanalytics import TextAnalyticsClient
from azure.core.credentials import AzureKeyCredential

# 🔐 Your Azure credentials
key = "EeN7RzBLgc7PnV2Bpyw3eCa3fsH4QLPN7dCvqoZtAp1T3mLVeuQNJQQJ99CDACGhslBXJ3w3AAAaACOGHJxg"
endpoint = "https://manish-ds-work.cognitiveservices.azure.com/"

# Create client
client = TextAnalyticsClient(endpoint=endpoint, credential=AzureKeyCredential(key))

# Multiple texts (batch)
documents = [
    "The product quality is amazing but delivery was very slow.",
    "Customer support was very helpful and resolved my issue quickly.",
    "The app crashes again and again. Very frustrating experience.",
    "Average service, nothing special."
]

# Analyze sentiment
response = client.analyze_sentiment(documents)

# Print results
for i, doc in enumerate(response, start=1):
    print(f"\nText {i}:")
    print("Sentiment:", doc.sentiment)
    print("Positive Score:", round(doc.confidence_scores.positive, 2))
    print("Neutral Score:", round(doc.confidence_scores.neutral, 2))
    print("Negative Score:", round(doc.confidence_scores.negative, 2))
```
